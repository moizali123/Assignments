```python
# Question 7

eng = int(input('Enter your English Marks:'));
math = int(input('Enter your Maths Marks:'));
urdu = int(input('Enter your Urdu Marks:'));
sci = int(input('Enter your Science Marks:'));
isl = int(input('Enter your Islamiat Marks:'));
total = 500;
obTotal = eng+math+urdu+sci+isl;
percent = (obTotal/total)*100;
if percent<=0:
    print("Error. No zero or negative percentage");
elif percent>=100:
    print('Error. percentage should not be 100 or greter than 100');
elif percent<100 and percent>=80:
    print('Your Grade is A+');
elif percent<80 and percent>=70:
    print('Your Grade is A');
elif percent<70 and percent>=60:
    print('Your Grade is B');
elif percent<60 and percent>=50:
    print('Your Grade is C');
elif percent<50 and percent>=40:
    print('Your Grade is D');
elif percent<40 and percent>=33:
    print('Your Grade is E');
elif percent<33:
    print('You are Fail. No Grade');
else:
    print('Error. Inappropriate inputs');
```
